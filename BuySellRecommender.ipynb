```python
!pip install crewai==0.28.8 crewai_tools==0.1.6 langchain_community==0.0.29
# Warning control
import warnings
warnings.filterwarnings('ignore')
from crewai import Agent, Task, Crew
import os
import openai
openai_api_key_from_user = input("Please enter your key for using OpenAI API: ")
tone_for_support = input("Please enter what tone you would want the support agent to use: e.g. friendly, assertive, professional")
company_to_research = input("URL for company's earnings report:")
financial_info_url = input("URL for financial info about the stock")
target_language = input("What language would you like the response to be in?")
customer_inquiry = input("Please enter your inquiry")


openai_api_key = openai_api_key_from_user
os.environ["OPENAI_API_KEY"] = openai_api_key
os.environ["OPENAI_MODEL_NAME"] = 'gpt-3.5-turbo'

investment_research_agent = Agent(
    role="Senior Investment Research Agent",
	goal="Be the most diligent financial researcher that reads every single detail"
        " about an assigned stock, note pros and cons carefully from an investor's perspective.",
	backstory=(
		"You are researching {company_to_research} and "
        " are now working on providing "
		"thorough and thoughtful pros and cons of investing money for next 2 years in that company."
		"You need to make sure that you don't miss out on any details but also focus on the big themes and big picture for helping make a buy/sell recommendation."
	),
	allow_delegation=False,
	verbose=True
)

buy_sell_recommendation_agent = Agent(
	role="Single stock investment advisor",
	goal="Get recognition for providing the "
    "best buy, hold or sell recommendation for a given stock",
	backstory=(
		"You are researching {company_to_research} with a view to make a buy or sell or hold recommendation and "
        "are now working with your team "
		"providing the best support possible.\n"
		"You need to work on the inputs from senior investment research agent."
	),
	verbose=True
)



from crewai_tools import SerperDevTool, \
                         ScrapeWebsiteTool, \
                         WebsiteSearchTool
docs_scrape_tool = ScrapeWebsiteTool(
    website_url=financial_info_url
)
investment_research = Task(
    description=(
    "{company_to_research} has been sekected to be researched on. \n"

		"Make sure to use everything you know "
        "to provide the best inputs possible."
		"You must strive to understand all aspects from future financial projections and possibilities of money growth or loss \n"
    "All such relevant themes along with details must be recorded."
    ),
    expected_output=(
	    "A detailed, informative response to the "
        "company being researched from all product, services and any other aspects "
        "that are relevant to understand future potential growth or decline of the company.\n"
        "The response should include references "
        "to everything you used to find the answer, "
        "including external data or solutions. "
        "Ensure the answer is complete, "
		"leaving no questions unanswered."
    ),
	tools=[docs_scrape_tool],
    agent=investment_research_agent,
)
buysell_recommendations = Task(
    description=(
        "Review the response drafted by the Senior Investment research agent for  {company_to_research}. "
        "Based on the response from the senior investment research agenet, make sure "
		"you provide a high quality buy or hold or sell recommendation.\n"
        "leaves no questions unanswered."
    ),
    expected_output=(
        "A final, detailed, and informative response "
        "ready to be used for making an investment decision.\n"
        ),
    agent=buy_sell_recommendation_agent,
)
crew = Crew(
  agents=[investment_research_agent, buy_sell_recommendation_agent],
  tasks=[investment_research, buysell_recommendations],
  verbose=2,
  memory=True
)
inputs = {
    "company_to_research": company_to_research
}
result = crew.kickoff(inputs=inputs)
#result = "just simple result"
# Extract the translated text from the API response

from IPython.display import Markdown
Markdown(result)
```
